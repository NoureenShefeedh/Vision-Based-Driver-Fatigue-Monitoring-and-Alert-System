```python
import os
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import shutil
import random
from tqdm import tqdm
```

```python
BASE_PATH = r"C:\Users\Noureen\OneDrive\Documents\DL\Vision-Based-Driver-Fatigue-Monitoring-and-Alert-System\DATA\archive\Driver Drowsiness Dataset (DDD)"
OUTPUT_PATH = os.path.join(BASE_PATH, "DATA_SPLIT")

classes = ["Drowsy", "Non Drowsy"]
```

```python
for split in ["train", "val", "test"]:
    for cls in classes:
        os.makedirs(os.path.join(OUTPUT_PATH, split, cls), exist_ok=True)
```

```python
def split_data(class_name, train_ratio=0.7, val_ratio=0.15):
    src_folder = os.path.join(BASE_PATH, class_name)
    images = os.listdir(src_folder)
    random.shuffle(images)

    total = len(images)
    train_end = int(total * train_ratio)
    val_end = train_end + int(total * val_ratio)

    splits = {
        "train": images[:train_end],
        "val": images[train_end:val_end],
        "test": images[val_end:]
    }

    for split, files in splits.items():
        for file in tqdm(files, desc=f"{class_name} → {split}"):
            src = os.path.join(src_folder, file)
            dst = os.path.join(OUTPUT_PATH, split, class_name, file)
            shutil.copy(src, dst)
```

```python
for cls in classes:
    split_data(cls)

print("✅ Dataset successfully split into Train / Val / Test!")
```

```python
for split in ["train", "val", "test"]:
    print(f"\n{split.upper()}")
    for cls in classes:
        path = os.path.join(OUTPUT_PATH, split, cls)
        print(f"{cls}: {len(os.listdir(path))}")
```
